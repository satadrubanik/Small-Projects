 # Netflix Campaign Design Generator
## Leveraging OpenAI API & Gradio UI for Creative Visual Content

**Project Overview:**
This notebook demonstrates the integration of OpenAI's `gpt-image-1` model with Gradio's web interface to generate bespoke banner and poster designs for Netflix's digital marketing campaigns. By transforming abstract text prompts into visually compelling images, this platform showcases how AI can revolutionize the creative design industry.

**Libraries Used:**
- `gradio` — Web interface for ML models
- `requests` — HTTP requests for fetching image data
- `PIL` (Pillow) — Image processing
- `io.BytesIO` — Binary data stream management
- `openai` — OpenAI API access
- `base64` — Decoding base64 image data

Cell 2 — Code: Import Libraries

```python
# Import essential libraries
import gradio as gr
import requests
from PIL import Image
from io import BytesIO
from openai import OpenAI
import base64

print("✅ All libraries imported successfully!")
```

Cell 3: Initialize OpenAI Client (Code)

```python
# Initialize the OpenAI client
# IMPORTANT: Replace with your actual OpenAI API key
client = OpenAI(api_key="YOUR_OPENAI_API_KEY_HERE")

print("OpenAI client initialized.")
```

Cell 4: Define the Image Generation Function (Code)

```python
def generate_image(prompt):
    """
    Generates a bespoke image from a text prompt using OpenAI's API.
    """
    try:
        # Call OpenAI's image generation API
        response = client.images.generate(
            model="dall-e-3", 
            prompt=prompt,
            size="1024x1024",
            quality="high",
            n=1
        )
        
        image_data = response.data[0]
        
        # Handle response formats (base64 or URL)
        if hasattr(image_data, 'b64_json') and image_data.b64_json:
            image_bytes = base64.b64decode(image_data.b64_json)
            image = Image.open(BytesIO(image_bytes))
        elif hasattr(image_data, 'url') and image_data.url:
            image_response = requests.get(image_data.url)
            image = Image.open(BytesIO(image_response.content))
        else:
            raise ValueError("No valid image data received from API.")
        
        return image
    
    except Exception as e:
        raise gr.Error(f"Error generating image: {str(e)}")
```

Cell 5: Create and Launch Gradio Interface (Code)

```python
# Create the Gradio interface
interface = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(
        label="Enter Your Design Prompt",
        placeholder="e.g., A cinematic Netflix banner for a sci-fi thriller set in deep space...",
        lines=3
    ),
    outputs=gr.Image(label="Generated Design", type="pil"),
    title="Netflix Campaign Design Generator",
    description="Transform abstract ideas into visually compelling designs for Netflix campaigns.",
    examples=[
        ["A cinematic Netflix banner for a sci-fi thriller set in deep space, dark moody atmosphere"],
        ["A dark and mysterious poster for a Netflix crime thriller, neon lights reflecting on wet pavement"],
        ["A vibrant and colorful banner for a Netflix comedy show, friends laughing in a cozy apartment"]
    ],
    flagging_mode="never"
)

# Launch the interface
interface.launch(share=True)
```
